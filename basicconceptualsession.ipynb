```python

numbers = input().split()

even_sum = 0
even_count = 0

for num in numbers:
    num = int(num)
    if num % 2 == 0:
        even_sum += num
        even_count += 1

print("Sum of even numbers:", even_sum)
print("Count of even numbers:", even_count)
     
```

```python

numbers = list(map(int, input().split()))

even_sum = 0
even_count = 0

for num in numbers:
    if num % 2 == 0:
        even_sum+=num
        even_count += 1

print("Sum of even numbers:", even_sum)
print("Count of even numbers:", even_count)
```
