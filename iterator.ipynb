```python
#bsic iterator example
set1={12,1,2,34,5}
number_iter=iter(set1)
print(type(number_iter))
print(next(number_iter)) #   মানে এর ভিতরের element গুলার কোনো fixed order নাই .tai output এ যেকোনো element আসতে পারে
print(next(number_iter))
print(next(number_iter))
print(next(number_iter))
print(next(number_iter))
 

```

```python
lst=[x for x in range(500)]
 
def numbers(chunk_size,lst):
    for i in range(len(lst)):
        yield lst[i:i+chunk_size]
sayed=numbers(5,lst)  
print(next(sayed))      
print(sayed )# generator object
```

```python
lst=[x for x in range(500)]
 
def numbers(chunk_size,lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]
sayed=numbers(5,lst) 
print(next(sayed))
print(next(sayed))
print(next(sayed))
print(next(sayed))
print(next(sayed))

```
