```python
# read file
with open("./sample_data/emails.txt", "r") as file:
    lines = file.readlines()

# cleaning
emails = list(map(lambda x: x.strip(), lines))
#lower
emails=list(map(lambda x:x.lower(),emails))

# remove duplicate set banai same email remove hoise
unique_emails = list(set(emails))


# sort
unique_emails.sort()

# write to new file
with open("./sample_data/unique_emails.txt", "w") as file:
    for email in unique_emails:
        file.write(email + "\n")

#output:
#akob@gmail.com
#sakuivb@gmail.com
#sheikhmohammedsaif@gmail.com
#sheikhmohammedsayed@gmail.com
    
```

```python
with open("./sample_data/names.txt", "r") as file:
    lines = file.readlines()

# cleaning
names = list(map(lambda x: x.strip().capitalize(), lines))

# remove duplicate
unique_names = list(set(names))

# sort
unique_names.sort()

# save
with open("./sample_data/clean_names.txt", "w") as file:
    for name in unique_names:
        file.write(name + "\n")
        
#output:
#Akib
#Arnsu
#Sakinnun
#Sauy
#Sayed       
```

```python
Write a function read_lines(filename) that reads a text file and prints each line with its line number.

📥 Sample Input (file: data.txt):
Apple
Banana
Cherry
📤 Sample Output:
1: Apple
2: Banana
3: Cherry
```

```python
def  read_lines(filename):
  with open("./sample_data/data.txt","r") as file:
    result=file.readlines()
    line_breakof=list(map(lambda x:x.strip(),result))
    line_number=1
    for i in line_breakof:
      print(f'{line_number}:{i}')
      line_number+=1

 
  
read_lines("data.txt")

#output
#1:Apple
#2:Banana
#3:Cherry
```

```python
Write a function read_from_position(filename, position) that uses the file pointer to start reading a file from a given byte position and prints the remaining content.

📥 Sample Input:
filename = "hello.txt"   # contains: "Hello, World!"
position = 7
📤 Sample Output:
World!


```

```python
def read_from_position(filename, position):
    with open("./sample_data/hello.txt", "r") as file:
        file.seek(position)   # cursor move kore
        data = file.read()    # read from that position
        print(data)

read_from_position("hello.txt", 7)
#output:
#World!   amr output e show korbe j ata 
```

```python
Write a function safe_read(filename) that tries to open and read a file. Use try, except, else, and finally blocks:
Print file content if successful (else block)
Handle the case where file does not exist (except block)
Always print "Operation complete." at the end (finally block)

📥 Sample Input 1:
safe_read("missing.txt")
📤 Sample Output 1:
Error: File not found.
Operation complete.
📥 Sample Input 2:
safe_read("hello.txt")   # contains: "Hello, World!"
📤 Sample Output 2:
Hello, World!
Operation complete.
```

```python
def safe_read(filename):
    try:
        with open("./sample_data/missing.txt", "r") as file:  #  apadoto dore ni file ta nai
            data = file.read()      # file open + read করার চেষ্টা
            
    except:
        print("Error: File not found.") # file না থাকলে error handle

    else:
        print(data)       #  যদি error না হয় → content print

    finally:
        print("Operation complete.")  #  সবসময় run হবে (important ❗)


safe_read("missing.txt")
```
