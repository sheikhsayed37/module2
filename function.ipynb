```python
def sayed(**kwargs):
    for keys,values in kwargs.items():#. kwargs.items() কি করে?
                                        #.items() দিলে dictionary থেকে key + value pair পাওয়া যায়
        print(f'{keys} :{values}')

sayed(a=2,b=3,n_name=50,c_name=40)
```
