```python
sq_add = lambda x, y: x**2 + y**2
print(sq_add(2, 3))
```

```python
add = lambda a, b: a + b
print(add(5, 10))   # Output: 15
```

```python
check_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even(5))
print(check_even(8))
```

```python
check_even= lambda x:"even" if x%2==0 else "odd"
print(check_even(6))
```

```python
operations = [
lambda x: x + 1,
lambda x: x * 2,
lambda x: x ** 2
]

for func in operations:
    print(func(5))# Output: 6, 10, 25
```

```python
math_ops = {
    'add': lambda x, y: x + y,
    'multiply': lambda x, y: x * y,
    'power': lambda x, y: x ** y
}

print(math_ops['add'](10, 5))    # Output: 15
print(math_ops['power'](2, 3))   # Output: 8
```

```python
#map
lst2=[1,2,3,4,5]
lst1=[1,2,8,7,7]
conv=map(lambda x,y:x**2+y**2,lst1,lst2)
print(list(conv))
```

```python
number=[1,2,3,4,5]
no=map(lambda x:x**2,number)
print(list(no))
```

```python
#filter
students = [
{'name': 'Alice', 'age': 20, 'grade': 'A'},
 {'name': 'Bob', 'age': 22, 'grade': 'B'},
{'name': 'Charlie', 'age': 19, 'grade': 'A'},
{'name': 'David', 'age': 21, 'grade': 'C'}
]

# 'A' গ্রেড এবং বয়স < ২১
top_young_students = filter(lambda s: s['grade'] == 'A' and s['age'] < 21, students)
print(list(top_young_students))
```

```python

lst = [1, 2, -3, 4, -5]

even_number = list(filter(lambda x: x % 2 == 0, lst))

print(even_number)
                  
                  
```

```python
সহজভাবে:
filter() → condition check করে (True হলে রাখে)
map() → value পরিবর্তন করে
```
