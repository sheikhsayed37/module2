```python
#filter
students = [
{'name': 'Alice', 'age': 20, 'grade': 'A'},
 {'name': 'Bob', 'age': 22, 'grade': 'B'},
{'name': 'Charlie', 'age': 19, 'grade': 'A'},
{'name': 'David', 'age': 21, 'grade': 'C'}
]

# 'A' গ্রেড এবং বয়স < ২১
top_young_students = filter(lambda s: s['grade'] == 'A' and s['age'] < 21, students)
print(list(top_young_students))
```

```python
সহজভাবে:
filter() → condition check করে (True হলে রাখে)
map() → value পরিবর্তন করেv
এক লাইনে বুঝো:
filter() 👉 "যেগুলো লাগে শুধু সেগুলো রাখো"
map() 👉 "সবগুলার উপর কিছু করো"
🎯 Real-life example:

ধরো:

filter() = gatekeeper 🚪 (ভালো হলে ঢুকতে দেয়)
map() = machine ⚙️ (সবাইকে process করে)
```

```python

from functools import reduce

lst = [1,2,3,4,5]

result = reduce(lambda x, y: x + y, lst)  #doita perameter x,y


print(result) # or return kora answer tao deko aktai o kinto pora kajer base e akta output dekaise
#reduce() → একটা value দেয় (list না)
```

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

# Lambda function ব্যবহার করে
summation = reduce(lambda x, y: x + y, numbers)
print(summation)# Output: 21
```

```python
Step 1: (1 + 2) → 3
Step 2: (3 + 3) → 6
Step 3: (6 + 4) → 10
Step 4: (10 + 5) → 15
Step 5: (15 + 6) → 21
Final Result: 21
```

```python
from functools import reduce
numbers = [10, 5, 25, 8, 30, 15]

max_value = reduce(lambda x, y: x if x > y else y, numbers)

print(max_value)     # Output: 30 pore list er basis e kaj korse
```

```python
words = ["Hello", "World", "Python", "Programming"]

sentence = reduce(lambda x, y: x + " " + y, words)

print(sentence)# Output: Hello World Python Programming
```
