```python
lst=[x for x in range (9)]

def batch_generator(chunk_size,lst):
    for i in range(1,len(lst),chunk_size):
        yield lst[i:i+chunk_size]
s1=batch_generator(3,lst)
print(next(s1))
print(next(s1))
print(next(s1))

```
