```python
#multiple inheritance

class cooling_mechnism:
   def __init__(self,colling_method):
      self.colling_method=colling_method

       
   def colling_method(self):                 # ai method ta create na korleo code same kaj korbe
      print(f"collnig process is gone a be {self.colling_method}")


class smartphone_with_coolinmood(smartphone,cooling_mechnism):
   def __init__(self,model,battery,camera,processor,colling_method):
      smartphone.__init__ (self,model,battery,camera,processor) 
      cooling_mechnism.__init__(self,colling_method)     


pro_mobile=smartphone_with_coolinmood("saifali",2333,70,"sd","sodium")


print(pro_mobile.model)
print(pro_mobile.processor)
print(pro_mobile.battery)
print(pro_mobile.colling_method)
```

```python
# parent class
class device:
    def __init__(self, brand, power):
        self.brand = brand
        self.power = power

    def start(self):
        print("Device is starting")


# single inheritance
class laptop(device):
    def __init__(self, brand, power, ram):
        super().__init__(brand, power)
        self.ram = ram

    def start(self):
        super().start()   # parent method call
        print("Laptop is ready")


# another class
class gaming_feature:
    def __init__(self, gpu):
        self.gpu = gpu

    def play_game(self):
        print(f"Game running with {self.gpu}")


# multiple inheritance
class gaming_laptop(laptop, gaming_feature):
    def __init__(self, brand, power, ram, gpu):
        laptop.__init__(self, brand, power, ram)
        gaming_feature.__init__(self, gpu)


# object
g1 = gaming_laptop("Dell", "65W", "16GB", "RTX 4060")

# output
g1.start()
g1.play_game()

print(g1.brand)
print(g1.ram)
print(g1.gpu)
```
