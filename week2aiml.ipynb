```python
# parent class
class device:
    def __init__(self, brand, power):
        self.brand = brand
        self.power = power

    def start(self):
        print("Device is starting")


# single inheritance
class laptop(device):
    def __init__(self, brand, power, ram):
        super().__init__(brand, power)
        self.ram = ram

    def start(self):
        super().start()   # parent method call
        print("Laptop is ready")

# another class
class gaming_feature:
    def __init__(self, gpu):
        self.gpu = gpu

    def play_game(self):
        print(f"Game running with {self.gpu}")


# multiple inheritance
class gaming_laptop(laptop, gaming_feature):
    def __init__(self, brand, power, ram, gpu):
        laptop.__init__(self, brand, power, ram)
        gaming_feature.__init__(self, gpu)

# object
g1 = gaming_laptop("Dell", "65W", "16GB", "RTX 4060")

# output
g1.start()
g1.play_game()

print(g1.brand)
print(g1.ram)
print(g1.gpu)
```

```python
# parent class
class car:
    def __init__(self, brand, engine, fuel):
        self.brand = brand
        self.engine = engine
        self.fuel = fuel

    def start(self):
        print("Car is starting")


# single inheritance
class electric_car(car):
    def __init__(self, brand, engine, fuel, battery_capacity):
        super().__init__(brand, engine, fuel)
        self.battery_capacity = battery_capacity

    def start(self):
        super().start()   # parent call
        print("Electric system ready")


# another class
class autopilot_system:
    def __init__(self, version):
        self.version = version

    def auto_drive(self):
        print(f"Autopilot version {self.version} driving")


# multiple inheritance
class tesla(electric_car, autopilot_system):
    def __init__(self, brand, engine, fuel, battery_capacity, version):
        electric_car.__init__(self, brand, engine, fuel, battery_capacity)
        autopilot_system.__init__(self, version)


# object create
t1 = tesla("Tesla", "Dual Motor", "Electric", "100kWh", "v12")

# output
t1.start() # jeheto sby akn tesla te dokse ga tai previous jekono method call korle pai jabo tbe o kinto tesla er first def e takte prbena
t1.auto_drive()

print(t1.brand)
print(t1.battery_capacity)
print(t1.version)
```

```python
class Camera:
  def __init__(self , name):
    self.name = name

 # method
  def capture(self):
    print("a photo is captured")
    
class Smart_Phone(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  #method overriding
  def capture(self):
    print("Photo is captured by a Phone")


class DSLR(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  def capture(self):
    print("Photo is captured  by DSLR")



class Drone(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  def capture(self):
    print("Photo is captured by Drone")



phone = Smart_Phone("Phone",30)
dslr = DSLR ("DSLR",200)
drone = Drone("Drone",150)


phone.capture()
dslr.capture()
drone.capture()
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei   # private variable

    def charge(self):
        print("phone is charging")

    def imei_getter(self): # imei acces korar jonno .jeta kina already privte
        return self.__imei

    def imei_setter(self,new_imei): # acces kore emei change korar jonno
        self.__imei=new_imei


iphone = Mobile("IPHONE", 12, "13diwiew")
print(iphone.imei_getter())
iphone.imei_setter("123456789")
print(iphone.imei_getter())
```
