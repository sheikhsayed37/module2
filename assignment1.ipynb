```python

Given a list of integers, extract all numbers that are divisible by 3 but not divisible by 5.

Input Format:
Space-separated integers

Output Format:
A list of numbers satisfying the condition

Example:
Input:
3 5 9 10 15 18 20 21

Output:
[3, 9, 18, 21]
```

```python
lst=input().split()

convert_to_integer=[int(num) for num in lst]

result=list(filter(lambda x:x%3==0 and x%5!=0,convert_to_integer))  # jeheto ata sorto based mane and ase seketre map dile true false return korbe 
          # per person teke r filter dile jara sorto poron korse tar ta applicable hbe


print(result)
```

```python
lst=input().split()

convert_to_integer=[int(num) for num in lst]

result=list(map(lambda x:x%3==0 and x%5!=0,convert_to_integer))  # jeheto ata sorto based mane and ase seketre map dile true false return korbe 
                                         # per person teke r filter dile jara sorto poron korse tar ta applicable hbe


print(result)
```

```python
Write a function compute(a, b) that:
Returns (a² + b²) / (a - b)
Use exception handling to manage:
Division by zero

Input Format:
Two values

Output Format:
Computed result OR error message
```

```python
def compute(a, b):
    try:
        result = (a**2 + b**2) / (a - b)
        return result
    except:
        print("Division by zero")
    else:
        print(result)
compute(1,1)




   
   
```

```python
Write a Python program that reads a text file and finds the most frequently occurring word.

Requirements:
Open and read data from a file (Example: data.txt)
Ignore case sensitivity ("Python" and "python" as the same)
Ignore punctuation marks
If multiple words have the same highest frequency, print any one of them

Input Format:
A text file containing multiple lines of text

Output Format:
The most frequent word

Example File Content (data.txt):
Python is easy to learn.
Python is powerful and Python is popular.

Output:
python
```

```python
with open("./sample_data/data.txt", "r") as file:
    text = file.read().lower()
    words=text.split()
    word_freq={}
    for i in words:
        if i in  word_freq:
            word_freq[i]= word_freq[i]+1
        else:
          word_freq[i]=1

# max ber korbo

max_word = ""
max_count = 0

for word in word_freq:
    if word_freq[word] > max_count:
        max_count = word_freq[word]
        max_word = word

print(max_word)
# print(max(word_freq, key=word_freq.get))
```

```python
Write a program that converts a sentence into a list representing the length of each word.Ignore extra spaces

Input Format:

A single line sentence

OutputFormat:
A list of integers representing word lengths

Example:
Input:
Machine learning is fun

Output:
[7, 8, 2, 3]

```

```python
string=input().split()

#print(type(string))
#print(string)
result=list(map(lambda x:len(x),string))

print(result)
```
